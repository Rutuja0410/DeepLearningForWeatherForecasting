```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df = pd.read_csv(csv_path)
df
```

```python
df = df[5::6]    #getting every hour data
df
```

```python
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
# df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')

df[:26]
```

```python
temp = df['T (degC)']
temp.plot()
```

```python
# Using last 5 hours temp to predict temp for current hour
def df_to_X_y(df, window_size=5):
    df_as_np = df.to_numpy()
    x = []
    y = []
    for i in range(len(df_as_np)-window_size):
        row = [[a] for a in df_as_np[i:i+window_size]]
        x.append(row)
        label = df_as_np[i+window_size]
        y.append(label)
    return np.array(x), np.array(y)

WINDOW_SIZE = 5
x, y = df_to_X_y(temp, WINDOW_SIZE)
print("x shape: ",x.shape, "y shape: ", y.shape)
```

```python
x_train, y_train = x[:60000], y[:60000]
x_val, y_val = x[60000:65000], y[60000:65000]
x_test, y_test = x[65000:], y[65000:]
x_train.shape, x_val.shape, x_test.shape
```

## LSTM Model

### Model Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8,'relu'))
model1.add(Dense(1,'linear'))


model1.summary()
```

```python
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001),metrics=[RootMeanSquaredError()])
```

### Model training

```python
model1.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=10)
```

### Model Prediction

```python
# Prediction on Validation data
val_predictions = model1.predict(x_val).flatten()
val_results = pd.DataFrame(data={'Val Predictions':val_predictions, 'Actuals':y_val})
val_results
```

```python
import matplotlib.pyplot as plt
plt.plot(val_results['Val Predictions'][:100])
plt.plot(val_results['Actuals'][:100])
```

```python
# Prediction on Test data
test_predictions = model1.predict(x_test).flatten()
test_results = pd.DataFrame(data={'Test Predictions':test_predictions, 'Actuals':y_test})
test_results
```

```python
plt.plot(test_results['Test Predictions'][:100])
plt.plot(test_results['Actuals'][:100])
```

### Model evaluation

```python
from sklearn.metrics import mean_squared_error as mse

def plot_predictions1(model, X, y, start=0, end=100):
    predictions = model.predict(X).flatten()
    df = pd.DataFrame(data={'Predictions': predictions, 'Actuals':y, 'Difference':(predictions-y)})
    plt.figure(figsize=(12, 6))
    plt.plot(df['Actuals'][start:end], label="True Values", color="blue")
    plt.plot(df['Predictions'][start:end], label="Predicted Values", color="orange", linestyle="--")
    plt.title("True vs Predicted Values")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.legend()
    plt.grid()
    plt.show()
    return df, mse(predictions, y)
```

```python
res, mserror = plot_predictions1(model1, x_test, y_test)
print('MSE :',mserror)
```

#### From the graph, we can infer that the difference between the actual and predicted values are relatively small, indicating that the model's predictions closely align with the true values.
